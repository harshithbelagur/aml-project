```python
# To use the new dataset, we're going to have to reconcile the player ids in this dataset with the player ids from the original dataset.

mod_player_info_df = common_player_info_df[['first_name', 'last_name', 'person_id']]
mod_player_info_df['player'] = mod_player_info_df.apply(lambda row: row['first_name'] + ' ' + row['last_name'], axis=1)
mod_player_info_df.drop(columns=['last_name', 'first_name'], inplace=True)


new_pid_df = player_pergame_df[['player', 'player_id']]

old_to_new_id_df = pd.merge(mod_player_info_df, new_pid_df, on = 'player').drop_duplicates().reset_index()

old_to_new_id_dict = old_to_new_id_df.set_index('person_id')['player_id'].to_dict()

```

```python
def old_to_new_ids(row, home_team=True):
  col = 'home_team_players' if home_team else 'away_team_players'
  new_ids = set()
  for old_id in row[col]:
    if old_id in old_to_new_id_dict:
      new_ids.add(old_to_new_id_dict[old_id])
  return new_ids

new_custom_df = pd.DataFrame(custom_df)

# now, we replace the old player ids in player_game_df with the new player ids
new_custom_df['home_team_players'] = custom_df.apply(old_to_new_ids, axis=1)
new_custom_df['away_team_players'] = custom_df.apply(lambda row: old_to_new_ids(row, False), axis=1)


new_custom_df.drop_duplicates(subset=['game_id'], keep='first', inplace=True)

new_custom_df
```

```python
# Now, we will see if we can gain any insight from OBPM (Offsensive Box Plus/Minus) and DBPM. OBPM is an advanced statistic used in the NBA to evaluate a player's offensive impact on the game.
# It takes into account how many points per 100 possesions a player contributes to their team's offense compared to the average nba player, normalizing
# for the number of minutes the player played. Note that DBPM is very similar, except it's for Defense.

# As a crude baseline, we'll just add up the the total OBPM for the players for the opposing teams for each game. We would expect that on average, the team with players that
# have a greater OBPM will have a greater chance of winning.

advanced_player_df =  pd.read_csv("/content/drive/MyDrive/aml-data/Advanced.csv")
advanced_player_df.set_index('player_id')

def team_obpm_sum(row, home=True):
  col = 'home_team_players' if home else 'away_team_players'
  # Note that we are only taking the top-8 players bc we assume that the others don't
  # get to play
  return sum(sorted([advanced_player_df.iloc[pid]['obpm'] for pid in row[col]])[0:8])

new_custom_df['home_obpm_sum'] = new_custom_df.apply(team_obpm_sum, axis=1)

new_custom_df['away_obpm_sum'] = new_custom_df.apply(lambda row: team_obpm_sum(row, False), axis=1)


def compare_obpm(row):
    if row['home_obpm_sum'] > row['away_obpm_sum']:
        return 'Home OBPM > Away OBPM'
    else:
        return 'Home OBPM <= Away OBPM'

new_custom_df['obpm_comparison'] = new_custom_df.apply(compare_obpm, axis=1)

plt.figure(figsize=(8, 6))
sns.barplot(x='obpm_comparison', y='wl_home', data=new_custom_df, ci=None)

plt.xlabel('Comparison between Home OBPM and Away OBPM')
plt.ylabel('Mean of wl_home')
plt.title('Correlation between Home OBPM and Win/Loss at Home')

plt.show()
```

```python
#hmmm that was kind of unexpected. When home OBPM is greater than away OBPM, the home team wins less? Maybe there's
# a different story for DBPM ....

def team_dbpm_sum(row, home=True):
    col = 'home_team_players' if home else 'away_team_players'
    # Note that we are only taking the top-8 players because we assume that the others don't
    # get to play
    return sum(sorted([advanced_player_df.iloc[pid]['dbpm'] for pid in row[col]])[0:8])

new_custom_df['home_dbpm_sum'] = new_custom_df.apply(lambda row: team_dbpm_sum(row), axis=1)
new_custom_df['away_dbpm_sum'] = new_custom_df.apply(lambda row: team_dbpm_sum(row, False), axis=1)

def compare_dbpm(row):
    if row['home_dbpm_sum'] > row['away_dbpm_sum']:
        return 'Home DBPM > Away DBPM'
    else:
        return 'Home DBPM <= Away DBPM'

new_custom_df['dbpm_comparison'] = new_custom_df.apply(compare_dbpm, axis=1)

plt.figure(figsize=(8, 6))
sns.barplot(x='dbpm_comparison', y='wl_home', data=new_custom_df, ci=None)

plt.xlabel('Comparison between Home DBPM and Away DBPM')
plt.ylabel('Mean of wl_home')
plt.title('Correlation between Home DBPM and Win/Loss at Home')

plt.show()

```

```python
# The sum of DBPM for the home team being larger than that of the away team helps the home team.

# To summarize, we found that a team's OBPM being larger than their oponent's hurts their chances at home, while a team's DBPM being larger
# than their opponent's helps. Apparently, defense is important for home games. Still, the result for OBPM is surprising and needs to be
# investigated further.
```

```python
# general team stats by season

team_summaries_df = pd.read_csv("/content/drive/MyDrive/aml-data/Team Summaries.csv")


# In addition to player-stats, the dataset also offers team-stats. Here, we will look at the some the "opponent stats per 100 possessions data", which
# details the opposing team's performance when playing a given team. We are going to try to find relationships between certain columns in this data and
# season-wide statistics, like overall win-loss ratio.

opponent_team_100_df =  pd.read_csv("/content/drive/MyDrive/aml-data/Opponent Stats Per 100 Poss.csv")
opponent_team_100_df.columns
```

```python
# Below, we are going to examine relationships between a team's opponent's statistics per 100 possessions, and a team's W-L stats for the regular season.
# For some stats, such as opp_tov_per_100_poss, we expect a postive correlation with wins. For other stats, such as oppp_pts_per_100_poss, we expect a negative correlation.

# Calculate win-loss ratio
team_summaries_df['win_loss_ratio'] = team_summaries_df['w'] / team_summaries_df['l']

# Filter data for the 2020 season and merge on 'team' column
team_summaries_df_2020_merged = pd.merge(team_summaries_df[team_summaries_df['season'] == 2020], opponent_team_100_df[team_summaries_df['season'] == 2020], on='team')

# List of opponent stat columns
opponent_stat_columns = ['opp_fg_per_100_poss', 'opp_tov_per_100_poss', 'opp_stl_per_100_poss', 'opp_trb_per_100_poss']

# Plot each opponent stat in its own graph
for column in opponent_stat_columns:
    plt.figure(figsize=(10, 8))
    plt.scatter(team_summaries_df_2020_merged[column], team_summaries_df_2020_merged['win_loss_ratio'])
    plt.title(f'Correlation between {column} and Win-Loss Ratio in 2020')
    plt.xlabel(column)
    plt.ylabel('Win-Loss Ratio')
    plt.grid(True)
    plt.show()
```

```python
team_summaries_df['win_loss_ratio'] = team_summaries_df['w'] / team_summaries_df['l']


team_summaries_df_2020_merged = pd.merge(team_summaries_df[team_summaries_df['season'] == 2020], opponent_team_100_df[team_summaries_df['season'] == 2020], on='team')

opponent_stat_columns = {
    'opp_fg_per_100_poss': 'Opponent Field Goals per 100 Possessions',
    'opp_tov_per_100_poss': 'Opponent Turnovers per 100 Possessions',
    'opp_stl_per_100_poss': 'Opponent Steals per 100 Possessions',
    'opp_trb_per_100_poss': 'Opponent Total Rebounds per 100 Possessions'
}

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.subplots_adjust(hspace=0.5, wspace=0.3)

for i, (column, label) in enumerate(opponent_stat_columns.items()):
    row = i // 2
    col = i % 2

    axes[row, col].scatter(team_summaries_df_2020_merged[column], team_summaries_df_2020_merged['win_loss_ratio'], label='Data Points')

    x = team_summaries_df_2020_merged[column].values.reshape(-1, 1)
    y = team_summaries_df_2020_merged['win_loss_ratio'].values.reshape(-1, 1)
    model = LinearRegression()
    model.fit(x, y)
    y_pred = model.predict(x)
    axes[row, col].plot(x, y_pred, color='red', label='Line of Best Fit')

    axes[row, col].set_title(f'{label}\nand Win-Loss Ratio in 2020')
    axes[row, col].set_xlabel(label)
    axes[row, col].set_ylabel('Win-Loss Ratio')
    axes[row, col].grid(True)
    axes[row, col].legend()

plt.tight_layout()
plt.show()
```
